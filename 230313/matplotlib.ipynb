### 9) 틱 설정 - xticks(),yticks()

```python
# 플롯이나 차트에서 축상의 위치 표시 지점 : 틱(tick)
# 틱에 써진 숫자 혹은 글자 : 틱 라벨(tick label)
# 틱의 위치나 틱 라벨 수동설정 : xticks()이나 yticks()을 사용
```

```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X,C)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

```python
# 틱 라벨 문자열에는 $$ 사이에 수학 문자식(Mathmatical expressions)을 
# 넣을 수도 있다.
#https://matplotlib.org/3.1.1/tutorials/text/mathtext.html
#코사인식에  -파이부터 +파이까지의 숫자에 대한 y값을 구한 예제

X = np.linspace(-np.pi, np.pi, 256) #파이
C = np.cos(X) #코사인
plt.title("TeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, np.pi / 2, 0, np.pi /2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
#         파이 기호로 표시하라
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

### 10) 여러개의 선 그리기

#### 1. 하나의 plot() 이용

```python
# 라인 플롯에는 선을 하나가 아니라 여러개를 그리고 싶은 경우
# x데이터, y데이터, 스타일 문자열을 반복하여 인수로 넘긴다.

# 하나의 선을 그릴 때처럼 x 데이터나 스타일 문자열을 생략할 수 없다!

t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t **3, 'g^-')
plt.show()
```

#### 2. 복수의 plot() 이용

```python
#하나의 plot()이 아니라 복수의 plot()을 하나의 그림에 겹쳐서 그릴 수도 있다.

plt.title("복수의 plot명령을 한 그림에서 표현")
style1 = {"c" : "b" , "lw" : 5, "ls" :"--", 
          "marker" : "o","ms" : 15,
          "mec" :"g","mew" :5,"mfc":"r"}
style2 = {"c" : "k" , "lw" : 3, "ls" :":", 
          "marker" : "s","ms" : 10,
          "mec" :"m","mew" :5,"mfc":"c"}

plt.plot([1,4,9,16],**style1) #* ->dict을 키워드 변수로 넘겨줄 때
plt.plot([9,16,4,1],**style2)

plt.show()
```

# Axes

```python
#Figure 객체 안에 여러 개의 플롯을 그리드(grid) 형태로 보여야하는 경우
#Figure 안에 있는 각각의 플롯은 Axes라고 불리는 객체에 속함

#Figure 안에 Axes를 생성하려면 원래 subplot()을 사용하여 명시적으로 Axes 객체를 얻어야 합니다. 
#그러나 plot()을 바로 사용해도 자동으로 Figure, Axes 객체를 생성해 줍니다.
```

```python
# subplot()은 그리드(grid) 형태의 Axes 객체들을 생성
# Figure가 행렬(matrix)이고 Axes가 행렬의 원소라고 생각하면 됨

# 예를 들어 위와 아래 두 개의 플롯이 있는 경우 행이 2 이고 열이 1인 2x1 행렬
# subplot 명령은 세개의 인수를 가지는데 처음 두개의 원소가 전체 그리드 행렬의 모양을
# 지시하는 두 숫자이고, 세번째 인수가 네 개 중 어느것인지를 의미하는 숫자

ax1 = plt.subplot(2, 1, 1)
# 여기에서 윗부분에 그릴 플롯 명령 실행

ax2 = plt.subplot(2, 1, 2)
# 여기세어 아랫부분에 그릴 플롯 명령 실행
```

```python
# 위/아래 두 개의 플롯을 하나의 Figure 안에 그리려면 다음처럼 명령을 실행해야 한다.
# 여기에서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫번째 플롯을
# 가리키는 숫자가 0이 아니라 1임에 주의해야 합니다.

# tight_layout() 명령을 실행하면 플롯 간의 간격을 자동으로 맞춰준다.

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) #row,column, 순서
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() #플롯 간의 간격 맞춤
plt.show()
```

```python
# 만약 2x2 형태의 네 개의 플롯이라면 다음과 같이 그린다.
# 이 때 subplot 의 인수는 (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시할 수도 있다. 
# Axes의 순서는 콘솔출력 순서와 동일(좌에서 우로, 위에서 아래로).

np.random.seed(0)

plt.subplot(221) #2,2,1
plt.plot(np.random.rand(5))
plt.title('axes 1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes 2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes 3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes 4')

plt.tight_layout()
plt.show()
```

```python
#subplots()에 인수를 두개만 전달하면 복수의 Axes 객체를 동시에 생성 가능
#2차원 ndarray 형태로 Axes 객체가 반환
#위의 것을 다르게 표현한 것

fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title("axes 1")  
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title("axes 2")  
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title("axes 3")  
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title("axes 4")  

plt.tight_layout()
plt.show()
```

```python
# 데이터 플로팅을 위한 영역으로 Figure에 포함되는 Artis
# 2차원의 경우 Axis 객체를 두 개 포함
# 3차원의 경우 Axis 객체를 세 개 포함
# Axis는 ticks와 tick labels을 제공하며 
# 이는 Axes의 데이터 범위에 대한 척도가 됩니다.

#Axes와 Axis, #subplot 과 subplots구분!!!!!!!!!!!!!!!!!!!

# Axes 클래스와 그 메서드는 주요한 Entry point이며, 대부분의 플로팅 메서드가 정의되어 있다.

# Inheritance
# matplotlob.artist.Artist → matplotlib.axes._base._AxesBase → matplotlib.axes._axes.Axes
```

```python
# 하나의 Axes 객체는 두 개 이상의 Axis 객체를 가진다. 
# Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체

# Axis 객체는 ticks(축의 표시)과 ticklabels(눈금 레이블을 지정하는 문자열)에 대한
# 규모와 제한을 설정한다.

# ticks의 위치는 Locator 객체에 의해 결정되며, 눈금 레이블 문자열은 Formatter 객체에
# 의해 형식이 지정된다.

# Locator와 Formatter 객체를 올바르게 조합하면 매우 세밀하게 tick 위치와 labels을
# 제어할 수 있다.
```

```python
# 여러가지 플롯을 하나의 Axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든
# 경우가 있다. 
# 이 때는 다음처럼 twinx 명령으로 대해 복수의 y 축을 가진 플롯을 만들 수 있다. 
# twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 만듭니다.

fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label = 'y0')
ax0.set_ylabel('y0')
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g:', label = 'y1')
ax1.set_ylabel('y1')
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```

# 바 차트

```python
# x 데이터가 카테고리 값인 경우에는 bar 명령과 barh 명령으로 바 차트(bar chart)
# 시각화를 할 수 있다. 
# 가로 방향으로 바 차트를 그리려면 barh 명령을 사용

# 바 차트 작성시 주의점은 첫번째 인수인 left가 
#x축에서 바(bar)의 왼쪽 변의 위치를 나타냄

y = [2, 3, 1]
x = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
# 각 bar 마다 색상을 적용할 수 있다.
# 적용하고 싶은 색상을 데이터 순으로 리스트로 생성

# 생성한 리스트를 bar()의 color 키워드 인수로 전달

# 그러면 아래와 같은 색상을 갖는 바 차트를 생성 가능

y = [2, 3, 1]
x = ['가', '나', '다']
colors = ["red", "green", "blue"]

plt.title("Bar Chart")
plt.bar(x, y, color = colors)
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
# xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가할 수 있다.
   #오차를 위해
# 다음 코드에서 alpha는 투명도를 지정한다. 0이면 완전 투명, 1이면 완전 불투명

np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
colors = ['yellow', 'skyblue', 'pink','green']
plt.barh(people, performance, xerr = error, alpha = 0.4,color =colors )
plt.xlabel('x 라벨')
plt.show()
```

```python
# xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가할 수 있다.
   #오차를 위해
# 다음 코드에서 alpha는 투명도를 지정한다. 0이면 완전 투명, 1이면 완전 불투명

np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
colors = ['#BFC8D7', '#E2D2D2', '#E3E2B4','#A2B59F'] #헥사코드
plt.barh(people, performance, xerr = error, alpha = 1.0,color =colors )
plt.xlabel('x 라벨')
plt.show()
```

# 컨투어 플롯

```python
# 입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우에는 3차원 자료가 된다. 
# 3차원자료를 시각화하는 방법은 명암이 아닌 등고선(contour)을 사용하는 방법이다. 
# contour 혹은 contourf 명령을 사용한다. 
# contour는 등고선만 표시하고 contourf는 색깔로 표시한다. 
# 입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 한다.

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha = .75, cmap = 'jet')
plt.contour(XX, YY, ZZ, colors = 'black')
plt.show()
```

# 3D 서피스 플롯

```python
# 3차원 플롯은 등고선 플롯과 달리 Axes3D라는 3차원 전용 axes를 생성하여 입체적으로
# 표시
# plot_wireframe(), plot_surface() 명령을 사용한다.

from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX ** 2 + YY ** 2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D Surface Plot')
ax.plot_surface(XX, YY, ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```
