```python
import time

start_time = time.time() #시작하는 시간
test = []

#1,000,000번을 반복해서 list만듭니다.
for _ in range(1_000_000):
    tp = [_, _] #list
    test.append(tp)
# 끝나는 시간 -  시작하는 시간 
print(time.time() - start_time)
# 동작한 시간
```

```python
import time

start_time = time.time()
test = []

#1,000,000번을 반복해서 tuple만듭니다.
for _ in range(1_000_000):
    tp = (_, _) #tuple 
    test.append(tp)
# 끝나는 시간 -  시작하는 시간 
print(time.time() - start_time)
# 동작한 시간
```
